```python
import requests as req
from bs4 import BeautifulSoup as bs
import csv
hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
```

1. import requests as req, baris kode ini mengimport library requests dengan alias req. Library requests merupakan library yang digunakan untuk mengirim HTTP request ke website yang diinginkan.
2. from bs4 import BeautifulSoup as bs, Baris kode ini mengimport library BeautifulSoup dengan alias bs dari library bs4. Library BeautifulSoup merupakan library yang digunakan untuk memudahkan proses parsing HTML atau XML.
3. import csv, baris kode ini mengimport library csv. Library csv merupakan library yang digunakan untuk membaca dan menulis file CSV (Comma Separated Values).
4. hades = {‘user-agent’: ‘Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36’}, baris kode ini mengisi sebuah dictionary dengan nama hades. Dictionary tersebut berisi informasi user-agent yang akan digunakan saat mengirim HTTP request. Informasi user-agent ini bisa memberitahu website tentang perangkat yang digunakan saat mengirim request[2].

### **Detik.com**

#### Liga Inggris

```python
def scrape_detik_LigaInggris():
    # URL halaman pertama
    url = f'https://www.detik.com/search/searchnews?query=liga%20inggris&result_type=latest'
    response = req.get(url, headers=hades).text
    soup = bs(response, 'lxml')

    # Kontainer artikel
    articles = soup.find_all('article', class_='list-content__item')
    if not articles:
        print("Error: Artikel tidak ditemukan.")
        return

    # Membuka file CSV dan menulis header
    data_to_save = []
    data_to_save.append(['No', 'Judul', 'Link', 'Teks', 'Media', 'Kategori Berita', 'Label'])

    # Proses setiap artikel
    count = 1
    for article in articles:
        try:
            # Ambil judul (h3)
            title_tag = article.find('h3', class_='media__title')
            title = title_tag.text.strip() if title_tag else "Judul tidak ditemukan"

            # Ambil link ke artikel lengkap
            link_tag = title_tag.find('a') if title_tag else None
            link = link_tag['href'] if link_tag else "Link tidak ditemukan"

            # Ambil isi konten dari halaman artikel
            content = "Konten tidak ditemukan"
            if link != "Link tidak ditemukan":
                article_response = req.get(link, headers=hades).text
                article_soup = bs(article_response, 'lxml')
                content_container = article_soup.find_all('div', class_='detail__body-text itp_bodycontent')

                paragraphs = []
                for c in content_container:
                    for p in c.find_all('p'):
                        # Skip advertisement and "Scroll to continue" messages
                        if 'ADVERTISEMENT' in p.text or 'SCROLL TO CONTINUE WITH CONTENT' in p.text:
                            continue
                        paragraphs.append(p.text)

                content = ' '.join(paragraphs).replace('\n', '').strip()

            # Menambahkan data ke list
            data_to_save.append([count, title, link, content, 'Detik.com', 'Sepakbola', 'Liga Inggris'])

            # Menampilkan hasil di layar
            print(f"No: {count}")
            print(f"Judul: {title}")
            print(f"Link: {link}")
            print(f"Teks: {content[:200]}...")  # Menampilkan preview 200 karakter pertama
            print("-" * 50)

            count += 1

        except Exception as e:
            print(f'Error processing article: {e}')

    # Menyimpan hasil ke CSV
    with open('ligaInggrisDetik.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file, delimiter=',')
        writer.writerows(data_to_save)
        print("\nData berhasil disimpan ke ligaInggrisDetik.csv")

# Panggil fungsi
scrape_detik_LigaInggris()
```

```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import os
```

#### Liga Inggris

```python
def scrape_tempo_liga_inggris():
    # Inisialisasi opsi Chrome
    options = Options()
    options.add_argument('--headless')  # Jalankan Chrome tanpa GUI
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    # Atur path untuk chromedriver
    os.environ["PATH"] += ":/usr/lib/chromium-browser/chromedriver:/usr/lib/chromium-browser/"

    # Inisialisasi WebDriver
    driver = webdriver.Chrome(options=options)

    # URL halaman pertama untuk pencarian 'liga inggris'
    url = f'https://www.tempo.co/search?q=liga+inggris'
    driver.get(url)

    # Tunggu halaman selesai dimuat
    time.sleep(5)

    # Ambil HTML dari halaman yang dimuat
    html = driver.page_source
    soup = bs(html, 'lxml')

    # Kontainer artikel
    articles = soup.select('aside figure')
    if not articles:
        print("Error: Artikel tidak ditemukan.")
        driver.quit()
        return

    # Membuka file CSV dan menulis header
    data_to_save = []
    data_to_save.append(['No', 'Judul', 'Link', 'Teks', 'Media', 'Kategori Berita', 'Label'])

    # Proses setiap artikel
    count = 1
    for article in articles:
        try:
            # Ambil judul
            title_tag = article.find('figcaption').find('a')
            title = title_tag.text.strip() if title_tag else "Judul tidak ditemukan"

            # Ambil link ke artikel lengkap dan buat URL lengkap
            link = title_tag['href'] if title_tag else "Link tidak ditemukan"
            full_link = f'https://www.tempo.co{link}' if link and not link.startswith('http') else link

            # Ambil isi konten dari halaman artikel
            content = "Konten tidak ditemukan"
            if full_link != "Link tidak ditemukan":
                article_response = req.get(full_link).text
                article_soup = bs(article_response, 'lxml')

                # Menemukan kontainer artikel menggunakan selector yang lebih tepat
                content_container = article_soup.find_all('div', id='content-wrapper')  # Sesuaikan dengan selector yang tepat

                # Menggabungkan seluruh teks dari paragraf dalam elemen konten
                content = ''
                for element in content_container:
                    paragraphs = element.find_all('p')
                    for para in paragraphs:
                        content += para.get_text() + ' '  # Menambahkan teks paragraf dan spasi

                # Menghapus newline dan spasi ekstra di awal/akhir konten
                content = content.replace('\n', '').strip()

            # Menambahkan data ke list
            data_to_save.append([count, title, full_link, content, 'Tempo.co', 'Sepakbola', 'Liga Inggris'])

            # Menampilkan hasil di layar
            print(f"No: {count}")
            print(f"Judul: {title}")
            print(f"Link: {full_link}")
            print(f"Teks: {content[:200]}...")  # Menampilkan preview 200 karakter pertama
            print("-" * 50)

            count += 1

        except Exception as e:
            print(f'Error processing article: {e}')

    # Menyimpan hasil ke CSV
    with open('LigaInggrisTempo.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file, delimiter=',')
        writer.writerows(data_to_save)
        print("\nData berhasil disimpan ke LigaInggrisTempo.csv")

    # Tutup driver
    driver.quit()

# Panggil fungsi
scrape_tempo_liga_inggris()
```

## **1. Cleansing**

Menghapus angka, tanda baca, karakter khusus, mention, link, emotikon, dan spasi ganda.

```python
# Library
import re  # Untuk regular expression

# Tahap 1: Cleansing
def cleansing(text):
    # Ubah ke huruf kecil
    text = text.lower()

    # Hapus angka
    text = re.sub(r"\d+", "", text)

    # Hapus username yang diawali dengan '@'
    text = re.sub(r"@\w+", "", text)

    # Hapus link (URL)
    text = re.sub(r"http\S+|www.\S+", "", text)

    # Hapus tanda baca kecuali spasi dan huruf
    text = re.sub(r'[^\w\s]', ' ', text)

    # Hapus emotikon
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # Emotikon wajah
        u"\U0001F300-\U0001F5FF"  # Simbol dan Piktogram
        u"\U0001F680-\U0001F6FF"  # Transportasi dan Simbol
        u"\U0001F700-\U0001F77F"  # Simbol Misc
        u"\U0001F780-\U0001F7FF"  # Simbol Misc lagi
        u"\U0001F800-\U0001F8FF"  # Ekstra
        u"\U0001F900-\U0001F9FF"  # Ekstra lagi
        u"\U0001FA00-\U0001FA6F"  # Ekstra simbol
        u"\U00002702-\U000027B0"  # Simbol lain
        u"\U000024C2-\U0001F251"  # Ekstra karakter berguna
        "]+", flags=re.UNICODE
    )
    text = emoji_pattern.sub(r'', text)

    # Hapus extra spasi
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Terapkan fungsi cleansing dan cek hasilnya
df['cleaned_text'] = df['Teks'].apply(cleansing)
display(df[['Teks', 'cleaned_text']])
```

```python
from collections import Counter

# Menggabungkan seluruh teks untuk tokenisasi
all_text = ' '.join(train_df['cleaned_text'])

# Tokenisasi seluruh teks
tokens = word_tokenize(all_text)

# Menyusun bigrams atau trigrams (pasangan kata atau tiga kata berturut-turut)
bigrams = ngrams(tokens, 2)  # Bigrams untuk pasangan kata
trigrams = ngrams(tokens, 3)  # Trigrams untuk tiga kata berturut-turut

# Menghitung frekuensi bigrams dan trigrams
bigram_freq = Counter(bigrams)
trigram_freq = Counter(trigrams)

# Menampilkan 10 bigrams dan trigrams paling umum
print("Most common bigrams:", bigram_freq.most_common(20))
print("Most common trigrams:", trigram_freq.most_common(20))
```

### Tokenisasi

```python
# Daftar kata yang perlu dikecualikan
custom_stopwords = ['panjat tebing']

# Fungsi untuk tokenisasi dengan pengecualian frasa
def custom_tokenize(text):
    # Tokenisasi terlebih dahulu
    tokens = word_tokenize(text)

    # Gabungkan kembali 'panjat tebing' jika terpisah
    tokens = ['panjat tebing' if tokens[i] == 'panjat' and tokens[i+1] == 'tebing' else tokens[i]
              for i in range(len(tokens)) if i+1 < len(tokens)]

    return tokens

# Tokenisasi pada data train dan test
train_df_tokenized = train_df['cleaned_text'].apply(custom_tokenize)
val_df_tokenized = val_df['cleaned_text'].apply(custom_tokenize)
test_df_tokenized = test_df['cleaned_text'].apply(custom_tokenize)

# Tampilkan contoh hasil tokenisasi
print("Contoh tokenisasi pada data train:")
print(train_df_tokenized.tail())

print("\nContoh tokenisasi pada data test:")
print(test_df_tokenized.tail())

print("\nContoh tokenisasi pada data validasi:")
print(val_df_tokenized.tail())
```

### **Stemming**

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Membuat objek stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming pada list token dengan pengecualian untuk "panjat tebing"
def stemming_tokens_with_exclusion(tokens):
    # Gantilah "panjat tebing" menjadi token yang tidak akan disinggung oleh stemmer
    tokens = ['panjat_tebing' if token == 'panjat tebing' else token for token in tokens]

    # Lakukan stemming pada semua token
    stemmed_tokens = [stemmer.stem(word) for word in tokens]

    # Kembalikan "panjat_tebing" menjadi "panjat tebing"
    stemmed_tokens = [token.replace("panjat_tebing", "panjat tebing") for token in stemmed_tokens]

    return stemmed_tokens

# Terapkan stemming dengan pengecualian pada data train, val, dan test
train_stemmed = train_no_stopwords.apply(stemming_tokens_with_exclusion)
val_stemmed = val_no_stopwords.apply(stemming_tokens_with_exclusion)
test_stemmed = test_no_stopwords.apply(stemming_tokens_with_exclusion)
```

```python
!pip install stanza
import stanza

# Mengunduh model bahasa Indonesia untuk Stanza
stanza.download('id')

# Inisialisasi pipeline Stanza dengan model bahasa Indonesia
nlp = stanza.Pipeline('id')
```

```python
# Fungsi untuk lemmatization dengan pengecualian kata "panjat tebing"
def lemmatize_text_with_exclusion(text):
    # Gantilah "panjat tebing" menjadi string yang tidak akan diubah
    text = text.replace("panjat tebing", "panjat_tebing")

    # Proses teks dengan Stanza
    doc = nlp(text)

    # Ambil lemma dari setiap kata dalam teks
    lemmatized_tokens = [word.lemma for sent in doc.sentences for word in sent.words]

    # Kembalikan "panjat_tebing" sebagai "panjat tebing" lagi
    lemmatized_tokens = [token.replace("panjat_tebing", "panjat tebing") for token in lemmatized_tokens]

    return lemmatized_tokens

# Terapkan lemmatization dengan pengecualian pada data train, val, dan test
train_lemmatized = train_no_stopwords.apply(lambda x: lemmatize_text_with_exclusion(' '.join(x)))
val_lemmatized = val_no_stopwords.apply(lambda x: lemmatize_text_with_exclusion(' '.join(x)))
test_lemmatized = test_no_stopwords.apply(lambda x: lemmatize_text_with_exclusion(' '.join(x)))
```

#### Dataset PyTorch

```python
# Dataset class untuk tokenized data
class TextDataset(Dataset):
    # Inisialisasi dataset, menerima encodings (data yang sudah diproses menjadi token) dan labels (label target)
    def __init__(self, encodings, labels):
        self.encodings = encodings  # Menyimpan encodings/tokenized input
        self.labels = labels  # Menyimpan label (target klasifikasi)

    # Fungsi untuk mendapatkan jumlah data dalam dataset
    def __len__(self):
        return len(self.labels)  # Mengembalikan jumlah data sesuai dengan jumlah label

    # Fungsi untuk mengambil satu item (data dan label) berdasarkan indeks
    def __getitem__(self, idx):
        # Mengambil data yang telah ditokenisasi berdasarkan indeks
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        # Menambahkan label untuk data pada indeks yang sesuai
        item['labels'] = torch.tensor(self.labels[idx])
        return item  # Mengembalikan data dan label dalam bentuk tensor
```

### Model BertModel - IndoBert
- Dropout = 0.3
- Epoch = 5
- Learning Rate = 1e-6

```python
# Model BERT Classifier
# Membuat model klasifikasi dengan BERT
class BertClassifier(nn.Module):  # Definisi kelas model yang merupakan subclass dari nn.Module
    def __init__(self, dropout=0.3):  # Konstruktor kelas, dengan parameter dropout default 0.3
        super(BertClassifier, self).__init__()  # Memanggil konstruktor kelas induk nn.Module
        self.bert = BertModel.from_pretrained('indolem/indobert-base-uncased')  # Menggunakan IndoBERT sebagai model pretrained untuk bahasa Indonesia
        self.dropout = nn.Dropout(dropout)  # Menambahkan layer dropout untuk regularisasi, untuk menghindari overfitting
        self.linear = nn.Linear(768, 5)  # Layer linear (fully connected) dengan input 768 (output dari BERT) dan output 5 (jumlah kelas)
        self.relu = nn.ReLU()  # Fungsi aktivasi ReLU untuk meningkatkan non-linearitas dalam model

    def forward(self, input_ids, attention_mask):  # Fungsi forward untuk melakukan perhitungan melalui model
        # Mengambil output dari BERT: `input_ids` adalah ID token yang dihasilkan oleh tokenizer, dan `attention_mask` digunakan untuk mengabaikan padding
        _, pooled_output = self.bert(input_ids=input_ids, attention_mask=attention_mask, return_dict=False)

        # Melakukan dropout pada output yang dihasilkan oleh BERT untuk regularisasi
        dropout_output = self.dropout(pooled_output)

        # Menyambungkan output ke layer linear
        linear_output = self.linear(dropout_output)

        # Menambahkan fungsi aktivasi ReLU pada output linear untuk memperkenalkan non-linearitas
        final_layer = self.relu(linear_output)

        # Mengembalikan output akhir yang berupa hasil prediksi klasifikasi
        return final_layer
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

def train_and_evaluate(model, train_loader, val_loader, test_loader, learning_rate, epochs):
    use_cuda = torch.cuda.is_available()
    device = torch.device("cuda" if use_cuda else "cpu")

    criterion = nn.CrossEntropyLoss()
    optimizer = Adam(model.parameters(), lr=learning_rate)

    if use_cuda:
        model = model.cuda()
        criterion = criterion.cuda()

    for epoch_num in range(epochs):
        total_acc_train = 0
        total_loss_train = 0

        y_true_train = []
        y_pred_train = []

        model.train()
        for batch in tqdm(train_loader):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            output = model(input_ids, attention_mask)

            loss = criterion(output, labels)
            total_loss_train += loss.item()

            preds = output.argmax(dim=1)
            acc = (preds == labels).sum().item()
            total_acc_train += acc

            y_true_train.extend(labels.cpu().numpy())
            y_pred_train.extend(preds.cpu().numpy())

            model.zero_grad()
            loss.backward()
            optimizer.step()

        # Metrics for training set
        train_acc = total_acc_train / len(train_loader.dataset)
        train_precision = precision_score(y_true_train, y_pred_train, average='weighted')
        train_recall = recall_score(y_true_train, y_pred_train, average='weighted')
        train_f1 = f1_score(y_true_train, y_pred_train, average='weighted')

        total_acc_val = 0
        total_loss_val = 0

        y_true_val = []
        y_pred_val = []

        model.eval()
        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)

                output = model(input_ids, attention_mask)

                loss = criterion(output, labels)
                total_loss_val += loss.item()

                preds = output.argmax(dim=1)
                acc = (preds == labels).sum().item()
                total_acc_val += acc

                y_true_val.extend(labels.cpu().numpy())
                y_pred_val.extend(preds.cpu().numpy())

        # Metrics for validation set
        val_acc = total_acc_val / len(val_loader.dataset)
        val_precision = precision_score(y_true_val, y_pred_val, average='weighted')
        val_recall = recall_score(y_true_val, y_pred_val, average='weighted')
        val_f1 = f1_score(y_true_val, y_pred_val, average='weighted')

        print(
            f'Epoch: {epoch_num + 1} | Train Loss: {total_loss_train / len(train_loader):.3f} | Train Acc: {train_acc:.3f} | '
            f'Train Prec: {train_precision:.3f} | Train Recall: {train_recall:.3f} | Train F1: {train_f1:.3f} | '
            f'Val Loss: {total_loss_val / len(val_loader):.3f} | Val Acc: {val_acc:.3f} | '
            f'Val Prec: {val_precision:.3f} | Val Recall: {val_recall:.3f} | Val F1: {val_f1:.3f}'
        )

    # Evaluate on test set
    total_acc_test = 0
    total_loss_test = 0

    y_true_test = []
    y_pred_test = []

    model.eval()
    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            output = model(input_ids, attention_mask)

            loss = criterion(output, labels)
            total_loss_test += loss.item()

            preds = output.argmax(dim=1)
            acc = (preds == labels).sum().item()
            total_acc_test += acc

            y_true_test.extend(labels.cpu().numpy())
            y_pred_test.extend(preds.cpu().numpy())

    # Metrics for testing set
    test_acc = total_acc_test / len(test_loader.dataset)
    test_precision = precision_score(y_true_test, y_pred_test, average='weighted')
    test_recall = recall_score(y_true_test, y_pred_test, average='weighted')
    test_f1 = f1_score(y_true_test, y_pred_test, average='weighted')

    print("\nTesting Results:")
    print(
        f'Test Loss: {total_loss_test / len(test_loader):.3f} | Test Acc: {test_acc:.3f} | '
        f'Test Prec: {test_precision:.3f} | Test Recall: {test_recall:.3f} | Test F1: {test_f1:.3f}'
    )

EPOCHS = 5
LR = 1e-6
model_asli = BertClassifier()

# Panggil fungsi model
train_and_evaluate(model_asli, train_loader, val_loader, test_loader, LR, EPOCHS)
```

#### Tokenisasi

```python
# Inisialisasi tokenizer IndoBERT
tokenizer = BertTokenizer.from_pretrained('indolem/indobert-base-uncased')

# Fungsi untuk tokenisasi data
def tokenize_data(texts):
    return tokenizer(texts.tolist(), padding=True, truncation=True, max_length=512, return_tensors='pt')

# Tokenisasi data pelatihan dan pengujian
train_encodings2 = tokenize_data(df_train2['cleaned_text'])
test_encodings2 = tokenize_data(df_test2['cleaned_text'])
val_encodings2 = tokenize_data(df_val2['cleaned_text'])
```

#### Dataset PyTorch

```python
# Dataset class untuk tokenized data
class TextDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item
```

#### Model BertModel - IndoBert

```python
# Model BERT Classifier
class BertClassifier(nn.Module):
    def __init__(self, dropout=0.3):
        super(BertClassifier, self).__init__()
        self.bert = BertModel.from_pretrained('indolem/indobert-base-uncased')  # IndoBERT
        self.dropout = nn.Dropout(dropout)
        self.linear = nn.Linear(768, 4)  # 4 sesuai dengan jumlah kelas
        self.relu = nn.ReLU()

    def forward(self, input_ids, attention_mask):
        _, pooled_output = self.bert(input_ids=input_ids, attention_mask=attention_mask, return_dict=False)
        dropout_output = self.dropout(pooled_output)
        linear_output = self.linear(dropout_output)
        final_layer = self.relu(linear_output)
        return final_layer
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

def train_and_evaluate(model, train_loader, val_loader, test_loader, learning_rate, epochs):
    use_cuda = torch.cuda.is_available()
    device = torch.device("cuda" if use_cuda else "cpu")

    criterion = nn.CrossEntropyLoss()
    optimizer = Adam(model.parameters(), lr=learning_rate)

    if use_cuda:
        model = model.cuda()
        criterion = criterion.cuda()

    for epoch_num in range(epochs):
        total_acc_train = 0
        total_loss_train = 0

        y_true_train = []
        y_pred_train = []

        model.train()
        for batch in tqdm(train_loader):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            output = model(input_ids, attention_mask)

            loss = criterion(output, labels)
            total_loss_train += loss.item()

            preds = output.argmax(dim=1)
            acc = (preds == labels).sum().item()
            total_acc_train += acc

            y_true_train.extend(labels.cpu().numpy())
            y_pred_train.extend(preds.cpu().numpy())

            model.zero_grad()
            loss.backward()
            optimizer.step()

        # Metrics for training set
        train_acc = total_acc_train / len(train_loader.dataset)
        train_precision = precision_score(y_true_train, y_pred_train, average='weighted')
        train_recall = recall_score(y_true_train, y_pred_train, average='weighted')
        train_f1 = f1_score(y_true_train, y_pred_train, average='weighted')

        total_acc_val = 0
        total_loss_val = 0

        y_true_val = []
        y_pred_val = []

        model.eval()
        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)

                output = model(input_ids, attention_mask)

                loss = criterion(output, labels)
                total_loss_val += loss.item()

                preds = output.argmax(dim=1)
                acc = (preds == labels).sum().item()
                total_acc_val += acc

                y_true_val.extend(labels.cpu().numpy())
                y_pred_val.extend(preds.cpu().numpy())

        # Metrics for validation set
        val_acc = total_acc_val / len(val_loader.dataset)
        val_precision = precision_score(y_true_val, y_pred_val, average='weighted')
        val_recall = recall_score(y_true_val, y_pred_val, average='weighted')
        val_f1 = f1_score(y_true_val, y_pred_val, average='weighted')

        print(
            f'Epoch: {epoch_num + 1} | Train Loss: {total_loss_train / len(train_loader):.3f} | Train Acc: {train_acc:.3f} | '
            f'Train Prec: {train_precision:.3f} | Train Recall: {train_recall:.3f} | Train F1: {train_f1:.3f} | '
            f'Val Loss: {total_loss_val / len(val_loader):.3f} | Val Acc: {val_acc:.3f} | '
            f'Val Prec: {val_precision:.3f} | Val Recall: {val_recall:.3f} | Val F1: {val_f1:.3f}'
        )

    # Evaluate on test set
    total_acc_test = 0
    total_loss_test = 0

    y_true_test = []
    y_pred_test = []

    model.eval()
    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            output = model(input_ids, attention_mask)

            loss = criterion(output, labels)
            total_loss_test += loss.item()

            preds = output.argmax(dim=1)
            acc = (preds == labels).sum().item()
            total_acc_test += acc

            y_true_test.extend(labels.cpu().numpy())
            y_pred_test.extend(preds.cpu().numpy())

    # Metrics for testing set
    test_acc = total_acc_test / len(test_loader.dataset)
    test_precision = precision_score(y_true_test, y_pred_test, average='weighted')
    test_recall = recall_score(y_true_test, y_pred_test, average='weighted')
    test_f1 = f1_score(y_true_test, y_pred_test, average='weighted')

    print("\nTesting Results:")
    print(
        f'Test Loss: {total_loss_test / len(test_loader):.3f} | Test Acc: {test_acc:.3f} | '
        f'Test Prec: {test_precision:.3f} | Test Recall: {test_recall:.3f} | Test F1: {test_f1:.3f}'
    )

EPOCHS = 5
LR = 1e-6
model = BertClassifier()

train_and_evaluate(model, train_loader2, val_loader2, test_loader2, LR, EPOCHS)
```
